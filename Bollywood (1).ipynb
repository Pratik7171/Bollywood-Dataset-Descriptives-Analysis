# Which month of the year, maximum number movie releases are seen? (Note: Extract a new column called month from ReleaseDate column.)

```python
Bolyy_df['ReleaseDate'] = pd.to_datetime(Bolyy_df['ReleaseDate'], format='%d-%b-%y')
```

```python
Bolyy_df['Month'] = Bolyy_df['ReleaseDate'].dt.month_name()
```

```python
month_releases = Bolyy_df['Month'].value_counts()
```

```python
max_release_month = month_releases.index[0]
max_count = month_releases.iloc[0]
```

```python
print("Releases by Month:")
print(month_releases)
print(f"\nMonth with Maximum Releases: {max_release_month} with {max_count} movies.")
```
